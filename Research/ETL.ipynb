```python
## Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
## Load the Raw Dataset

df = pd.read_csv("dataset/raw/Telco-Customer-Churn.csv")
df.head()
```

# Section 2

TRANSFORM

```python
df.info()
```

```python
# Change TotalCharges Data Type
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Confirm change
df.info() 
```

```python
# Handle Missing Values
df.dropna(inplace=True)

# Confirm no missing values
df.isnull().sum()

```

```python

# Check and Remove Duplicates 
df.duplicated().sum()
```

```python
# Standardise Column Names
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()

list(df.columns)
```

```python
# Convert Binary Columns to Numerical
binary_cols = [
    "partner",
    "dependents",
    "phoneservice",
    "paperlessbilling",
    "churn"
]
df[binary_cols] = df[binary_cols].replace({"Yes": 1, "No": 0})

df[binary_cols].head()
```

```python
# Feature Engineering
df["tenure_group"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 60, 100],
    labels=["0-1yr", "1-2yr", "2-4yr", "4-5yr", "5+yr"]
)

df[["tenure", "tenure_group"]].head(10)
```

```python
# Final Data Check
df.info()
df.describe()
df.head()
```
